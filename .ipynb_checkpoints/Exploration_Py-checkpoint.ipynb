```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from statsmodels.graphics import tsaplots

sns.set(rc={"figure.figsize": (15, 6)})
sns.set_palette(sns.color_palette("Set2", 10))
sns.set_style("whitegrid")
```

```python
CO2_measurement = pd.read_csv('data/CO2_sensor_measurements.csv', sep='\t', index_col=0)
```

```python
choosen_location = 'AJGR'
```

```python
mask_location = CO2_measurement['LocationName'] == choosen_location
```

```python
time_serie = CO2_measurement[mask_location][['CO2']]
```

```python
time_serie.plot()
#plt.savefig('raw_data.svg')

time_serie_dif = time_serie.pct_change().dropna()
time_serie_dif.plot()
#plt.savefig('dif_data.svg')

tsaplots.plot_acf(time_serie, lags=100)
#plt.savefig('acf.svg')

tsaplots.plot_pacf(time_serie, lags=100)
#plt.savefig('pacf.svg')

plt.show()
```
